## 1 A 2-D Scatter Plot
* A scatter plot (also called a scatterplot, scatter graph, scatter chart, scattergram, or scatter diagram) is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data.
* If the points are color-coded, one additional variable can be displayed.
* The data are displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis and the value of the other variable determining the position on the vertical axis. (Wikipedia)

## 2 Cumulative Distribution Function (CDF)
### (Please refer explanation on CDF before continuing)
Definition of PDF:
* For a continuous function, the probability density function (pdf) is the probability that the variate has the value x. Since for continuous distributions the probability at a single point is zero, this is often expressed in terms of an integral between two points.

Definition of CDF
* The cumulative distribution function (cdf) is the probability that the variable takes a value less than or equal to x.

#### 2.1 CDF of Petal Length for Setosa

```python
hist, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10, density = True)
pdf = hist/(sum(hist))
cdf = np.cumsum(pdf)
plt.grid()
plt.plot(bin_edges[1: ],pdf);
plt.plot(bin_edges[1: ], cdf)
plt.xlabel("PL_Setosa")
plt.ylabel("Probability")
```

**Observations:**
* Reading PDF: Around 22% of Setosa have Petal Length Between 1.5 and 1.6
* Reading CDF: Around 82% of Setosa have Petal Length <= 1.6 (Not possible from PDF)
* CDF always starts from 0 (left bottom) to 1 (top right) which means all Setosa have PL <= 1.9

**Building a CDF:**

**Method 1:**
* Example if 41 out of 50 setosa have PL <= 1.6, then CDF(at 1.6) = 41/50 = 0.82
* This can be repeated for all the points between 0 and 1.9 using linspace to plot the CDF

**Method 2:**
* We know PDF at each point gives the percentage of flowers having PL at that point
* So a cumulative sum of PDF uptil a certain point will give the CDF at that point
* Basically, CDF at a point can be calculated by calculating the area under the curve of PDF uptil that point
* By Calculus we know area under the curve can be found using integration
* So integrating PDF gives CDF and so differentiating CDF gives PDF

#### 2.2 CDF of Petal Length for all the three species

```python
## setosa
hist, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10, density = True)
pdf = hist/(sum(hist))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1: ],pdf, 'r', label = 'setosa');
plt.plot(bin_edges[1: ], cdf, 'r')

## versicolor
hist, bin_edges = np.histogram(iris_versicolor['petal_length'], bins=10, density = True)
pdf = hist/(sum(hist))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1: ],pdf, 'g', label = 'versicolor');
plt.plot(bin_edges[1: ], cdf, 'g')

## virginica
hist, bin_edges = np.histogram(iris_virginica['petal_length'], bins=10, density = True)
pdf = hist/(sum(hist))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1: ],pdf, 'b', label = 'virginica');
plt.plot(bin_edges[1: ], cdf, 'b')

plt.xlabel("petal_length")
plt.ylabel("Probability")
plt.legend()
```

**Observations**
* if..else using PDF:
* if (PL <= 2) then setosa; else if (2 < PL <= 5) then versicolor; else virginica
* Now looking at CDF, around 96% of Versicolar have PL <= 5, so our rule can be expected to give 96% correct results in case of Versicolor
* Similarly we can see around 10% of Verginica flowers have PL <= 5 and hence we can expect a mistake of 10% in our rule for Verginica
* These information cannot be understood using PDF alone

## 3 Mean and Standard Deviation

#### 3.1 Mean

```python
print("Means:")
print(np.mean(iris_setosa["petal_length"]))
print(np.mean(iris_virginica["petal_length"]))
print(np.mean(iris_versicolor["petal_length"]))
```

**Observations:**
* Just looking at the means we can make out that setosa tends to have an average PL of about 1.464
* Similarly Virginica and Versicolor have an avg PL of 5.552 and 4.26 resp
* Conclusively Setosa tend to have smaller PL than Verginica and Versolor

#### Effect of an outlier on mean

```python
print(np.mean(np.append(iris_setosa["petal_length"],35)))
```

* Adding only one outlier (somebody could enter a wrong value) corrupts the PL mean by a lot of margin
* **Thus mean is very sensitive to ouliers**

#### 3.2 Standard Deviation

```python
print("\nStd-dev:")
print(np.std(iris_setosa["petal_length"]))
print(np.std(iris_virginica["petal_length"]))
print(np.std(iris_versicolor["petal_length"]))
```

## 4 Median, Percentile, Quantile, MAD

#### 4.1 Median

```python
print("\nMedians:")
print(np.median(iris_setosa["petal_length"]))
print(np.median(iris_virginica["petal_length"]))
print(np.median(iris_versicolor["petal_length"]))
```

** Effect of outlier on the median**

```python
print(np.median(np.append(iris_setosa["petal_length"],35)))
```

**Observations**
* First thing to note is median values are very similar to mean values
* Note that median is not affected much by adding an outlier - Why?
* For median to get corrupted there must be many outliers (more than 30 - 40%)

#### 4.2 Percentiles
A percentile (or a centile) is a measure used in statistics indicating the value below which a given percentage of observations in a group of observations fall

```python
print("\n90th Percentiles:")
print(np.percentile(iris_setosa["petal_length"],90))
print(np.percentile(iris_virginica["petal_length"],90))
print(np.percentile(iris_versicolor["petal_length"], 90))
```

**Observations**: This means:
* 90% of Setosa have Petal length <= 1.7
* 90% of Virginica have Petal length <= 6.31
* 90% of Versicolor have Petal length <= 4.8

Try with other values

#### 4.3 Quantiles
* 0th, 25th, 50th, 75th and 100th percentiles are called Quantiles - (Divide the data into quarters - 4 equal parts)
* The respective ranges are also called as first quantile, second quantile, third quantile and fourth quantile

```python
print("\nQuantiles:")
print(np.percentile(iris_setosa["petal_length"],np.arange(0, 101, 25)))
print(np.percentile(iris_virginica["petal_length"],np.arange(0, 101, 25)))
print(np.percentile(iris_versicolor["petal_length"], np.arange(0, 101, 25)))
```

**Observations:**

**Setosa:**
* 25% of Setosa have Petal Length <= 1.4
* 50% of Setosa have Petal Length <= 1.5
* 75% of Setosa have Petal Length <= 1.575

**What about Versicolor and Verginica?**

### Inter Quartile Range (IQR):
* IQR is the range where the middle 50% of the points lie
* IQR = (75th Percentile - 25th Percentile)

#### 4.4 Median Absolute Deviation (MAD)
* The median absolute deviation(MAD) is a robust measure of how spread out a set of data is.
* robust.mad() method from statsmodels computes MAD

```python
from statsmodels import robust
print ("\nMedian Absolute Deviation")
print(robust.mad(iris_setosa["petal_length"]))
print(robust.mad(iris_virginica["petal_length"]))
print(robust.mad(iris_versicolor["petal_length"]))
```

**Note:**
* Absolute deviation for a point 'a' = abs(a - median)
* MAD is the sum over absolute deviations over all points
* Just as Std deviation is the measure of avg deviation from the mean, MAD is the measure of avg deviation from the median
